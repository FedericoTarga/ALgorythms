PROBLEM: Given an array of integers. Find the Inversion Count in the array.

Inversion Count: For an array, inversion count indicates how far (or close) the array is from being sorted. If array is already sorted then the inversion count is 0. If an array is sorted in the reverse order then the inversion count is the maximum. Formally, two elements a[i] and a[j] form an inversion if a[i] > a[j] and i < j


EXAMPLE A: Input: N = 5, arr = [2, 4, 1, 3, 5] Output: 3 Explanation: The sequence 2, 4, 1, 3, 5 has three inversions (2, 1), (4, 1), (4, 3)

EXAMPLE B: Input: N = 5 , arr = [2, 3, 4, 5, 6] Output: 0 Explanation: As the sequence is already sorted so there is no inversion count.

EXAMPLE C: Input: N = 3, arr = [10, 10, 10] Output: 0 Explanation: As all the elements of array are same, so there is no inversion count.

and so on....


```python
class Solution:
    # Function to count inversions in the array.
    def inversionCount(self, arr, n):
        # Your Code Here
        def merge_sort(arr, temp, left, right):
            inv_count = 0

            if left < right:
                mid = (left + right) // 2
                inv_count += merge_sort(arr, temp, left, mid)
                inv_count += merge_sort(arr, temp, mid + 1, right)
                inv_count += merge(arr, temp, left, mid, right)

            return inv_count

        def merge(arr, temp, left, mid, right):
            i = left
            j = mid + 1
            k = left
            inv_count = 0

            while i <= mid and j <= right:
                if arr[i] <= arr[j]:
                    temp[k] = arr[i]
                    i += 1
                else:
                    temp[k] = arr[j]
                    j += 1
                    inv_count += (mid - i + 1)

                k += 1

            while i <= mid:
                temp[k] = arr[i]
                k += 1
                i += 1

            while j <= right:
                temp[k] = arr[j]
                k += 1
                j += 1

            for m in range(left, right + 1):
                arr[m] = temp[m]

            return inv_count

        temp = [0] * n
        return merge_sort(arr, temp, 0, n - 1)
```

```python
# Test function to validate the inversionCount function
def test_inversionCount():
    sol = Solution()
    test_cases = [
        ([2, 4, 1, 3, 5], 3),  #Inversions (2, 1), (4, 1), (4, 3).
        ([5, 4, 3, 2, 1], 10), #Sorted in reverse order, inversion count is maximum (n * (n - 1) / 2)
        ([2, 4, 1, 3, 5], 3),  #Inversions: (2, 1), (4, 1), (4, 3)
        ([10, 10, 10, 10], 0), #All elements are the same, no inversions
        ([1, 20, 6, 4, 5], 5), #Inversions: (20, 6), (20, 4), (20, 5), (6, 4), (6, 5)
    ]

    for arr, expected_count in test_cases:
        actual_count = sol.inversionCount(arr, len(arr))
        assert actual_count == expected_count, f"Error: Expected {expected_count}, but got {actual_count}."

    print("ALL TEST CASES PASSED \U0001F913!") 
# \U0001F913! is the escape Unicode for the smiling face with sunglasses
# Run the test function
test_inversionCount()
```
