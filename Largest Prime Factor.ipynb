PROBLEM:
Given a number N, the task is to find the largest prime factor of that number.

Exemple A) 
Input:
N = 5
Output:
5

Explanation:
5 has 1 prime factor i.e 5 only.

Exemple B)
Input:
N = 24
Output:
3

Explanation:
24 has 2 prime factors 2 and 3 in which 3 is greater.

Exemple C)
Input:
N = 56
Output:
7

Explanation:
56 has 2 prime factors 2 and 7 in which 7 is greater.

```python
class Solution:
    def largestPrimeFactor(self, N):
        # Initialize the largest prime factor to be 2
        largest_prime = 2
        
        # Divide N by 2 until it's no longer divisible by 2
        while N % 2 == 0:
            N //= 2
        
        # Now, try to divide N by odd numbers starting from 3
        # Increment by 2 to consider only odd numbers
        for i in range(3, int(N**0.5) + 1, 2):
            while N % i == 0:
                largest_prime = i
                N //= i
        
        # If N is a prime number greater than 2
        if N > 2:
            largest_prime = N
        
        return largest_prime


```

```python
# TEST 1 - See Example A above
N = 5
solution = Solution()
print('The largest prime factor of given number is:', solution.largestPrimeFactor(N))
```

Expanation:
--------------

This code defines a class Solution with a method largestPrimeFactor that takes an integer N as input and returns the largest prime factor of N. The code first divides N by 2 until it's no longer divisible by 2, and then it checks for prime factors by iterating through odd numbers up to the square root of N. Finally, if there's a remaining value of N greater than 2, it must be a prime factor, so it's assigned as the largest prime factor.

Steps:
-----

- Class Definition: The code defines a class called Solution.

- Method Definition: largestPrimeFactor: Inside the Solution class, there is a method called largestPrimeFactor which takes an integer N as an input parameter. This method will calculate and return the largest prime factor of N.

- Initializing Largest Prime Factor: At the beginning of the method, we initialize the largest_prime variable with 2. This is a safe starting point because 2 is the smallest prime number.

- Dividing by 2: We use a while loop to repeatedly divide N by 2 as long as it remains divisible by 2. This step effectively removes all the factors of 2 from N.

- Dividing by Odd Numbers: After handling the factor of 2, we move on to dividing N by odd numbers. We start a for loop that iterates over odd numbers, starting from 3 and going up to the square root of N. The use of odd numbers only is because even numbers greater than 2 are not prime.

- Checking for Divisibility: Inside the loop, we check if N is divisible by the current odd number i. If it is, we update largest_prime to the value of i (since i is a prime factor). We then update N by dividing it by i repeatedly until it's no longer divisible by i. This step effectively removes all occurrences of the prime factor i from N.

- Prime Factor Remaining: After the loop finishes, if N is still greater than 2, it means there's a prime factor greater than all the odd numbers we checked. In this case, we update largest_prime to the value of N.

- Returning the Result: Finally, we return the value of largest_prime, which is the largest prime factor of the original number N.
